## With BatchNorm

```python
import matplotlib.pyplot as plt
```

### Learning rate = 0.001, Epochs = 50

```python
test = []
train = []
train_loss = []

steps = []
step = 1

file = open('project-ai-scream/withbn_test/test0001epoch50.txt', 'r')

for line in file:
    line = line.split(',')
    line[2] = line[2].split(':')

    accuracy = float(line[2][1])
    test.append(accuracy)
    steps.append(step)
    step += 1

file = open('project-ai-scream/withbn_train/train0001epoch50.txt', 'r')

steps_t = []
step = 1

for line in file:
    line = line.split(',')
    line[2] = line[2].split(':')
    # print(line[2])
    line[3] = line[3].split(':')

    loss = float(line[2][1])
    accuracy = float(line[3][1])
    train_loss.append(loss)
    train.append(accuracy)
    steps_t.append(step)
    step += 1
```

```python
plt.plot(steps, test, label='test')
plt.plot(steps_t, train, label='train')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Number of Steps')
plt.legend()
plt.show()
```

```python
# plt.plot(steps, test, label='test')
plt.plot(steps_t, train_loss, label='train')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Loss vs Number of Steps')
plt.legend()
plt.show()
```
